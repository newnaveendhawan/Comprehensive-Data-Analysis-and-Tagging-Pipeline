```python
import pandas as pd
import re
import wordninja

# Preprocessing function to clean and split text
def format_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'\b(c/s|cs|cust\.|c\.s)\b', 'customer states', text)
    text = re.sub(r'\bck\b', 'check', text)
    text = re.sub(r'\bcust\b', 'customer', text)
    text = re.sub(r'[.:;\/-]', ' ', text)  # Replace special characters
    words = wordninja.split(text)
    return ' '.join(words)

# Keyword lists (prioritize longer phrases first)
components = [
    'steering wheel', 'heated steering wheel', 'steering column',
    'heater module', 'wiring harness', 'horn switch', 'trim cover',
    'airbag', 'module', 'harness', 'cover', 'stitching', 'connector',
    'button', 'coil', 'bezel', 'sensor'
]

failures = ['loose', 'not working', 'noise', 'failure', 'error', 'issue', 'fault', 'defective', 'broken', 'incorrect', 'malfunction', 'glitch', 'bug', 'downtime', 'disruption', 'crash', 'missing', 'inconsistent', 'unstable', 'unresponsive', 'damaged', 'mismatch', 'invalid', 'degraded', 'corrupted', 'blocked', 'jammed', 'overheating', 'leakage', 'short-circuit', 'misaligned', 'abnormal', 'disconnected', 'expired', 'unauthorized', 'incomplete', 'rejected', 'failed']


# Extract component and failure tags
def extract_tags(text):
    text = text.lower()
    comp_tag, fail_tag = None, None

    # Check components (longer phrases first)
    for comp in sorted(components, key=lambda x: len(x), reverse=True):
        if comp in text:
            comp_tag = comp.title()
            break  # Stop after first match

    # Check failures
    for fail in sorted(failures, key=lambda x: len(x), reverse=True):
        if fail in text:
            fail_tag = fail.title()
            break

    return comp_tag, fail_tag

# Load data
df = pd.read_excel('/content/DA -Task 2..xlsx')

# Preprocess columns
df['CUSTOMER_VERBATIM'] = df['CUSTOMER_VERBATIM'].apply(format_text)
df['CORRECTION_VERBATIM'] = df['CORRECTION_VERBATIM'].apply(format_text)

# Generate tags
df[['CUSTOMER_COMPONENT', 'CUSTOMER_FAILURE']] = df['CUSTOMER_VERBATIM'].apply(
    lambda x: pd.Series(extract_tags(x))
)
df[['CORRECTION_COMPONENT', 'CORRECTION_FAILURE']] = df['CORRECTION_VERBATIM'].apply(
    lambda x: pd.Series(extract_tags(x))
)

# Save results
df.to_excel('output_.xlsx', index=False)
print("Tags generated successfully!")
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from fuzzywuzzy import fuzz

# Download NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Load the Excel file
file_path = '/content/DA -Task 2..xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Predefined list of problems
problem_keywords = [
    "Heated", "Defected", "Battery issue", "Not working", "Switches issue", "Short-circuit", "Burnt", "Corroded",
    "Rusted", "Weak", "Fused", "Scratched", "Dented", "Torn", "Unresponsive", "Jammed", "Low power", "Dead",
    "Misaligned", "Discolored", "Worn out", "Glitch", "Unstable", "No signal", "Blurry", "Noisy", "Delayed", "Overloaded"
]

# Exclusion list of words to ignore
exclusion_list = [
    "okay","Torn", "Tor", "ed", "ng", "ming", "co", "pra", "w", "el", "pre", "n", "inop", "hrs", "st", "g", "cked", "sop", "r", "b",
    "la", "ok", "c", "h", "coming", "ro", "sts", "ne", "and", "the", "is", "to", "of", "for", "on", "with", "at", "by",
    "part", "Torn", "Tor", "tear"
]

# Preprocessing function
def extract_keywords(text):
    if isinstance(text, str):  # Ensure text is not NaN
        # Tokenize and remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = word_tokenize(text.lower())
        filtered_tokens = [
            word for word in tokens
            if word.isalnum()  # Alphanumeric check
            and word not in stop_words  # Exclude stopwords
            and len(word) <= 15  # Exclude words longer than 15 characters
            and len(word) >= 5   # Exclude words shorter than 4 characters
            and word.isalpha()   # Exclude words containing numeric characters
        ]

        # POS tagging to extract meaningful words
        pos_tags = nltk.pos_tag(filtered_tokens)
        keywords = [word for word, tag in pos_tags if tag.startswith(('NN', 'VB', 'JJ'))]  # Nouns, Verbs, Adjectives

        return list(set(keywords))  # Remove duplicates
    return []

# Function to normalize keywords
def normalize_keywords(keywords):
    normalized_keywords = []
    normalization_dict = {
        "heat": "Heated",
        "heater": "Heated",
        "not work": "Not working",
        "switch": "Switches issue",
        "short": "Short-circuit",
        "burn": "Burnt",
        "corrode": "Corroded",
        "rust": "Rusted",
        "weak": "Weak",
        "fuse": "Fused",
        "scratch": "Scratched",
        "dent": "Dented",
        "unresponsive": "Unresponsive",
        "jam": "Jammed",
        "low power": "Low power",
        "dead": "Dead",
        "misalign": "Misaligned",
        "discolor": "Discolored",
        "wear": "Worn out",
        "glitch": "Glitch",
        "unstable": "Unstable",
        "no signal": "No signal",
        "blur": "Blurry",
        "noise": "Noisy",
        "delay": "Delayed",
        "overload": "Overloaded"
    }
    for keyword in keywords:
        # Normalize using the dictionary
        for key, value in normalization_dict.items():
            if fuzz.partial_ratio(keyword, key) > 70:  # Use fuzzy matching for flexibility
                normalized_keywords.append(value)
                break
        else:
            # If no match is found, keep the original keyword
            normalized_keywords.append(keyword)
    return list(set(normalized_keywords))  # Remove duplicates

# Function to classify text into predefined problems
def classify_problem(text):
    matched_problems = []
    for problem in problem_keywords:
        if fuzz.partial_ratio(text, problem) > 80:  # Use fuzzy matching for flexibility
            matched_problems.append(problem)
    return matched_problems

# Enhanced keyword categorization
def categorize_keywords(text):
    # Extract keywords
    keywords = extract_keywords(text)
    # Normalize keywords
    keywords = normalize_keywords(keywords)
    # Classify problems
    problems = classify_problem(" ".join(keywords))
    return problems[:1]  # Limit to 1 main problem per category

# Create new columns for categorized keywords
df['Problem1'] = ''
df['Problem2'] = ''

# Populate the new columns with categorized keywords
for index, row in df.iterrows():
    # Extract keywords from CORRECTION_VERBATIM
    correction_text = row['CORRECTION_VERBATIM']
    problem1 = categorize_keywords(correction_text)

    # Write to Problem1 column
    df.at[index, 'Problem1'] = ', '.join(problem1) if problem1 else ''

    # Extract keywords from CUSTOMER_VERBATIM
    customer_text = row['CUSTOMER_VERBATIM']
    problem2 = categorize_keywords(customer_text)

    # Write to Problem2 column
    df.at[index, 'Problem2'] = ', '.join(problem2) if problem2 else ''

# Save the updated DataFrame to a new Excel file
output_file_path = 'Task2 with Tag problem1.xlsx'
df.to_excel(output_file_path, index=False)
print(f"Processed data saved to {output_file_path}")
```
